```python
import pandas as pd

# Load data (update path as needed)
df = pd.read_csv('/content/example_data.csv')

# Extract year and month from 'month' column
df['Year'] = df['month'].str.slice(0,4).astype(int)
df['Month'] = df['month'].str.slice(5,7).astype(int)

# Filter for years 2007-2019
df_filtered = df[(df['Year'] >= 2007) & (df['Year'] <= 2019)]

# Filter for total employment and both sexes
df_filtered = df_filtered[(df_filtered['variable'] == 'Employment') & (df_filtered['sex'] == 'Both sexes')]

# Select relevant columns (Year, Month, Alberta, Ontario)
df_filtered = df_filtered[['Year', 'Month', 'Alberta', 'Ontario']]

# Check cleaned data
print(df_filtered.head())
print(df_filtered.info())
print(df_filtered.describe())
```

```python
# Step 1: Install Required Libraries
# Reinstall torch and torchvision with CUDA support
!pip uninstall -y torch torchvision torchaudio
!pip install torch torchvision torchaudio --extra-index-url https://download.pytorch.org/whl/cu118 # Or replace cu118 with your CUDA version if different

# Install other libraries
!pip install -q transformers accelerate bitsandbytes

# %%
from transformers import AutoTokenizer, AutoModelForCausalLM, pipeline
import torch

model_id = "mistralai/Mistral-7B-Instruct-v0.1"  # Hugging Face LLM

# Load tokenizer and quantized model (4-bit to save GPU memory)
tokenizer = AutoTokenizer.from_pretrained(model_id)
model = AutoModelForCausalLM.from_pretrained(
    model_id,
    device_map="auto",
    load_in_4bit=True,
    torch_dtype=torch.float16
)

# Setup text generation pipeline
generator = pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
    max_new_tokens=1024,
    temperature=0.7,
    top_p=0.95,
    repetition_penalty=1.2
)
```

```python
# Generate stats from the data
avg_alberta = df["Alberta"].mean()
avg_ontario = df["Ontario"].mean()
start_year = df["Year"].min()
end_year = df["Year"].max()

summary = f"""
Between {start_year} and {end_year}, Alberta had an average employment of {avg_alberta:.1f}, while Ontario had an average of {avg_ontario:.1f}. This data reveals consistent trends in employment across both provinces with visible monthly shifts.
"""
print(summary)
```

```python
prompt = f"""
Write an engaging, SEO-optimized blog post for a Canadian staffing and recruitment company using this employment data:

{summary}

The blog should include:
- Key employment trends
- Regional comparison (Alberta vs Ontario)
- Practical tips for job seekers
- Insights for employers and recruiters
- Recommendations for staffing companies

Make the post interesting, data-driven, and formatted like a professional blog. The goal is to achieve 1,000 likes and 500+ comments on LinkedIn.
"""
```

```python
# Generate blog post
output = generator(prompt)[0]['generated_text']

# Display result
print("\n===== GENERATED BLOG POST =====\n")
print(output)
```

```python
# Define SEO keywords and their backlinks
backlink_map = {
    "Statistics Canada": "https://www.statcan.gc.ca",
    "remote work": "https://www.canada.ca/en/employment-social-development",
    "Ontario job market": "https://www.ontario.ca/page/labour-market",
    "Canadian staffing industry": "https://www.cacsc.com"
}

def add_backlinks(text, backlinks):
    for keyword, link in backlinks.items():
        if keyword in text:
            text = text.replace(keyword, f"[{keyword}]({link})")
    return text

# After blog is generated
raw_blog = generator(prompt)[0]["generated_text"]
final_blog = add_backlinks(raw_blog, backlink_map)

print(final_blog)
```
